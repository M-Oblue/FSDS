```python
cursor.execute('create table cardataset.car(buying VARCHAR(10),maint VARCHAR(10),doors VARCHAR(10),persons VARCHAR(10),lug_boot VARCHAR(10),safety VARCHAR(10),class VARCHAR(10))')
```

```python
import csv
with open('car.data','r') as c:
    car_data = csv.reader(c , delimiter='\n')
    query = """insert into cardataset.car (buying,maint,doors,persons,lug_boot,safety,class) values(%s,%s,%s,%s,%s,%s,%s)"""
    for i in car_data:
        split = str(i[0]).split(",")
        
        cursor.execute(query,split)
mydb.commit()
```

```python
# inserting data using placeholders(%s) instead of string formating(f'').

import csv
with open('car.data','r') as c:
    # without delimiter command data is already in ready format,split was required because we used delimiter.
    car_data = csv.reader(c)
    query = """insert into cardataset.car values(%s,%s,%s,%s,%s,%s,%s)"""
    for i in car_data:
        print(i)
        cursor.execute(query,i)
mydb.commit()
```

```python
# creating a tuple of read data to enter using {}.format()

import csv
with open('car.data','r') as c:
    # without delimiter command data is already in ready format,split was required because we used delimiter.
    car_data = csv.reader(c)
    for i in car_data:
        query = "insert into cardataset.car values{}".format(tuple(i))
        #row_data = tuple(i)
        cursor.execute(query)
        #print(i)
mydb.commit()
```
